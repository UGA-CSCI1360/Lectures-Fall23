### Part A (Decision Tree Model), 50 points

[athens_airbnb_preprocessed.csv](athens_airbnb_preprocessed.csv) is the ready dataset file after applying the preprocessing steps from Question 1. Use the file to train a decision tree model that satisfies the following requirements:

- 15% of the dataset records should be selected randomly for model testing.
- The maximum depth of the tree should be 3.
- Apply K-fold cross validation algorithm when training the model, where k = 5.
- The model's test accuracy average should exceed 65%.


#### Grading Rubric
- Correct file loading
- Correct split into X and y.
- Uses k-fold cross validation.
- Correct split into train/test parts.
- Mean test accuracy exceeds 65%.


```python
# import necessary modules
import numpy as np
from sklearn import tree
import pandas as pd

from sklearn.model_selection import train_test_split


df = pd.read_csv('athens_airbnb_preprocessed.csv', header=None)

df
```

```python
np_arr = df.to_numpy()

X = np_arr[:, :-1]
y = np_arr[:, -1]

print(X.shape)
print(y.shape)
```

```python
test_results = []
clf = tree.DecisionTreeClassifier(max_depth=3)

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(
    X, y,  test_size=0.15)
    clf = clf.fit(X_train, y_train)
    test_results.append(clf.score(X_test, y_test))

print('test accuracy', sum(test_results) / len(test_results))
```
