```python
import pandas as pd

import numpy as np

df = pd.read_csv('athens_airbnb.csv')

# del df['listing_date']

# df.to_csv('athens_weekends.csv', index=False)

df

# arr = np.loadtxt('athens_weekends.csv',delimiter=',',skiprows=0)  
```

```python
from sklearn.impute import SimpleImputer
np_arr = df.to_numpy()
imputer = SimpleImputer(strategy='most_frequent')
np_arr = imputer.fit_transform(np_arr)
# df
# df.to_csv('athens_weekends.csv', index=False)

# np_arr[np_arr > 5]
np_arr[139]
```

```python
from sklearn import preprocessing

encoder = preprocessing.OrdinalEncoder()

np_arr[:, 2] = encoder.fit_transform([np_arr[:, 2]])
np_arr[:, 3] = encoder.fit_transform([np_arr[:, 3]])
np_arr[:, 4] = encoder.fit_transform([np_arr[:, 4]])
np_arr[:, 6] = encoder.fit_transform([np_arr[:, 6]])

np_arr[139]

# np_arr[:, 2]
```

```python
# Standardize features

X = np_arr[:, 1:-1]

y = np_arr[:, -1]

print(X.shape)
print(y.shape)


X = preprocessing.StandardScaler().fit_transform(X)

# print(np_arr.shape)
# print(X.shape)
# y.shape
print(X[139])
print(y[139])
```

```python
print(len(y[y  == 1]))
print(len(y[y  == 0]))
```

```python
import random
items_to_add = len(y[y  == 1]) - len(y[y  == 0])
indices_of_minor_label = np.where(y == 0)[0]
for i in range(items_to_add):
    # select an item randomly
    random_id = random.randint(0, len(indices_of_minor_label) - 1)
    # print(indices_of_minor_label[random_id])
    X = np.append(X, [X[ indices_of_minor_label[random_id] ]], axis=0)
    y = np.append(y, y[ indices_of_minor_label[random_id] ])
    # print(y[ indices_of_minor_label[random_id] ])

# indices_of_minor_label
print(len(y[y  == 1]))
print(len(y[y  == 0]))
```

```python
print(X.shape)
print(y.shape)

y2 = np.reshape(y, (y.shape[0], 1))

print(y2.shape)

xy = np.append(X, y2, 1)
print(xy.shape)

np.savetxt("athens_airbnb_preprocessed.csv", xy, delimiter=",")
```
